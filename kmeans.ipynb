```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
```

# K-Means

### Generate Data

```python
def generate_cluster(n, d, center, sigma):
    return np.random.randn(n, d) * sigma + np.array(center)


# Create dummy dataset
n = 100
d = 2
X1 = generate_cluster(n, d, [[20, 0]], 10)
X2 = generate_cluster(n, d, [[-20, -40]], 10)
X3 = generate_cluster(n, d, [[-40, 20]], 10)

plt.scatter(X1[:,0], X1[:,1])
plt.scatter(X2[:,0], X2[:,1])
plt.scatter(X3[:,0], X3[:,1])

# shuffle data
X = np.vstack((X1, X2, X3))
p = np.arange(X.shape[0])
np.random.shuffle(p)
X = X[p, :]
```

### Initialization

```python
K = 3
maxit = 10

centers = np.random.randn(K, d)
```

### Loop

```python
for i in range(maxit):
    # compute distance
    dists = [np.linalg.norm(X - c, axis=1, keepdims=True) for c in centers]     # N x 1
    dists = np.hstack(dists)                                                    # N x k

    # assign class
    class_label = np.argmin(dists, axis=1)                                      # N
    sum_dist = 0
    for k in range(K):
        sum_dist += sum(dists[class_label == k, k])
    print(sum_dist / n)

    # update centers
    for k in range(K):
        X_k = X[class_label == k]
        if len(X_k) > 0:
            centers[k, :] = np.mean(X_k, axis=0)
        else:
            centers[k, :] = X[np.random.choice(X.shape[0])]

# return class_label
```
